```python
import pandas as pd
import numpy as np
import seaborn
import matplotlib.pyplot as plt
```

```python
filename ="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
df = pd.read_csv(filename, header = None)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
```

```python
find = df.eq('?')
#Tìm giá trị'?' có trong dataframe và trả về True nếu tìm thấy, gán giá trị True vào biến find
result = df[find.any(axis=1)]
#Kiểm tra giá trị True có trong biến find theo chiều ngang, trả về một series giá trị True tương ứng với giá trị ?df.r
result
```

```python
#Các cột có giá trị ? là normalized-losses, num-of-doors, bore, stroke, hoursepower, peak-rpm, price
```

```python
#Đối với các cột có dtypes là object thì sử dụng replace theo freq, dtypes là num nói chung thì sử dụng replace theo mean
```

```python
df.replace ('?', np.nan, inplace = True)
#Thay gia tri ? bang gia tri NaN
df.dropna(subset = ['price'], axis = 0, inplace = True)
#Xoa dong co gia tri NaN tai cot price
df.reset_index(inplace = True)
#Reset lai index sau khi xoa dong 
```

```python
avg_nor = df['normalized-losses'].astype(float).mean()
avg_bore = df['bore'].astype(float).mean()
avg_stroke = df['stroke'].astype(float).mean()
avg_power = df['horsepower'].astype(float).mean()
avg_peak = df['peak-rpm'].astype(float).mean()
#Doi dtypes cua cac cot va tinh trung binh cac cot
```

```python
df.replace(np.nan, avg_nor, inplace = True)
df.replace(np.nan, avg_bore, inplace = True)
df.replace(np.nan, avg_stroke, inplace = True)
df.replace(np.nan, avg_power, inplace = True)
df.replace(np.nan, avg_peak, inplace = True)
#Thay gia tri NaN bang gia tri avg
```

```python
df['num-of-doors'].value_counts()
#Kiem tra so lan xuat hien cua moi gia tri trong cot
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace = True)
#Thay gia tri NaN bang four do gia tri four xuat hien nhieu nhat
```

```python
df.dtypes
#Kiem tra data format
#Cac gia tri normalized-losses, bore, stroke, horsepower, peak-rpm, price phai la dang num
```

```python
df['normalized-losses'] = df['normalized-losses'].astype(float)
df['bore'] = df['bore'].astype(float)
df['stroke'] = df['stroke'].astype(float)
df['horsepower'] = df['horsepower'].astype(float)
df['peak-rpm'] = df['peak-rpm'].astype(float)
df['price'] = df['price'].astype(float)
#Doi data type
```

```python
df['city-L/100km'] = 235/df['city-mpg']
df['highway-L/100km'] = 235/df['highway-mpg']
#Chuyen data ve dang common format
df.drop('city-mpg',inplace = True, axis = 1)
df.drop('highway-mpg',inplace = True, axis = 1)
#Xoa cot
```

```python
df['normalized-losses']=df['normalized-losses']/df['normalized-losses'].max()
df['bore']=df['bore']/df['bore'].max()
df['stroke']=df['stroke']/df['stroke'].max()
df['horsepower']=df['horsepower']/df['horsepower'].max()
df['peak-rpm']=df['peak-rpm']/df['peak-rpm'].max()
df['price']=df['price']/df['price'].max()
df['height']=df['height']/df['height'].max()
df['width']=df['width']/df['width'].max()
df['length']=df['length']/df['length'].max()
df['symboling']=df['symboling']/df['symboling'].max()
df['wheel-base']=df['wheel-base']/df['wheel-base'].max()
df['curb-weight']=df['curb-weight']/df['curb-weight'].max()
df['engine-size']=df['engine-size']/df['engine-size'].max()
df['compression-ratio']=df['compression-ratio']/df['compression-ratio'].max()
df['city-L/100km']=df['city-L/100km']/df['city-L/100km'].max()
df['highway-L/100km']=df['highway-L/100km']/df['highway-L/100km'].max()
#Chuyen data dang num ve cung similar range tu 0 - 1, su dung phuong phap simple feature scalling
```

```python
df2 = pd.get_dummies(df['fuel-type'])
df3 = pd.get_dummies(df['aspiration'])
#Chuyen data phan loai dang object thanh data dang num
```

```python
df2.rename(columns = {'gas':'fuel-gas', 'diesel':'fuel-diesel'}, inplace = True)
df3.rename(columns = {'std':'aspiration-std', 'turbo':'aspiration-turbo'}, inplace = True)
#Doi ten cot
```

```python
df = pd.concat([df2,df], axis = 1)
df = pd.concat([df3,df], axis = 1)
#Nhap hai dataframe moi vao dataframe cu
```

```python
df.drop('fuel-type', axis = 1, inplace = True)
df.drop('aspiration', axis = 1, inplace = True)
#Xoa hai cot cu
```
