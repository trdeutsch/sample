```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/mac/Desktop/Project/titanic (1)/train.csv')
test = pd.read_csv('/Users/mac/Desktop/Project/titanic (1)/test.csv')
```

```python
data.head()
```

```python
data.drop(columns = ['Name', 'Ticket', 'Cabin'], inplace = True)
```

```python
data.isnull().sum()
```

```python
data['Age'].replace(np.nan, data['Age'].mean(), inplace = True)
data['Embarked'].replace(np.nan, 'S', inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
data['Sex'] = LabelEncoder().fit(data['Sex']).transform(data['Sex'])
data['Embarked'] = LabelEncoder().fit(data['Embarked']).transform(data['Embarked'])
```

```python
data.head()
```

```python
X = np.array(data[['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']])
Y = np.array(data['Survived'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
X_train_nor = StandardScaler().fit(x_train).transform(x_train.astype(float))
X_test_nor = StandardScaler().fit(x_test).transform(x_test.astype(float))
```

```python
print('Train shape', X_train_nor.shape, y_train.shape)
print('Test shape', X_test_nor.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

search = GridSearchCV(LogisticRegression(),{
    'solver' : ['saga', 'sag', 'lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    'penalty' : ['l1', 'l2', 'elasticnet'],
    'C' : [0.001, 0.01, 0.1]
})

search.fit(X_train_nor, y_train)
search.cv_results_
```

```python
df = pd.DataFrame(search.cv_results_)
```

```python
df[['param_C', 'param_penalty', 'param_solver', 'mean_test_score']].sort_values(by = 'mean_test_score', ascending = False)
```

```python
lr = LogisticRegression(C = 0.01, penalty = 'l2', solver = 'saga')
lr.fit(X_train_nor, y_train)
y_predict = lr.predict(X_test_nor)
print('accuracy_score', accuracy_score(y_test, y_predict))
```
