```python
#Function that return top 10 player in their position
def position(x):
    if x == 'GK':
        global condition1
        global df_GK
        condition1 = df['position'] == 'GK'
        df_GK = df[condition1]
        df_GK.sort_values(by = 'overall', ascending  = False, inplace = True)
        return df_GK.head(10)
    elif x == 'RB':
        global condition2
        global df_RB
        condition2 = df['position'] == 'RB'
        df_RB = df[condition2]
        df_RB.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_RB.head(10)
    elif x =='CB':
        global condition3
        global df_CB
        condition3 = df['position'] == 'CB'
        df_CB = df[condition3]
        df_CB.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_CB.head(10)
    elif x == 'LB':
        global condition4
        global df_LB
        condition4 = df['position'] == 'LB'
        df_LB = df[condition4]
        df_LB.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_LB.head(10)
    elif x == 'CM':
        global condition5
        global df_CM
        condition5 = df['position'] == 'CM'
        df_CM = df[condition5]
        df_CM.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_CM.head(10)
    elif x == 'CDM':
        global condition6
        global df_CDM
        condition6 = df['position'] == 'CDM'
        df_CDM = df[condition6]
        df_CDM.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_CDM.head(10)
    elif x == 'CAM':
        global condition7
        global df_CAM
        condition7 = df['position'] == 'CAM'
        df_CAM = df[condition7]
        df_CAM.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_CAM.head(10)
    elif x == 'RW':
        global condition8
        global df_RW
        condition8 = df['position'] == 'RW'
        df_RW = df[condition8]
        df_RW.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_RW.head(10)
    elif x == 'LW':
        global condition9
        global df_LW
        condition9 = df['position'] == 'LW'
        df_LW = df[condition9]
        df_LW.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_LW.head(10)
    elif x == 'ST':
        global condition10
        global df_ST
        condition10 = df['position'] == 'ST'
        df_ST = df[condition10]
        df_ST.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_ST.head(10)
    elif x == 'CF':
        global condition11
        global df_CF
        condition11 = df['position'] == 'CF'
        df_CF = df[condition11]
        df_CF.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_CF.head(10)
```

```python
#Function that return top 10 player in top 5 nationality that contributed the most in Fifa21
def nationality(x):
    if x == 'England':
        global cond
        global df_eng
        cond = df['nationality'] == 'England'
        df_eng = df[cond]
        df_eng.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_eng.head(10)
    elif x == 'Germany':
        global cond2 
        global df_ger
        cond2 = df['nationality'] == 'Germany'
        df_ger = df[cond2]
        df_ger.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_ger.head(10)
    elif x == 'Spain':
        global cond3 
        global df_spain
        cond3 = df['nationality'] == 'Spain'
        df_spain = df[cond3]
        df_spain.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_spain.head(10)
    elif x == 'Argentina':
        global cond4 
        global df_argentina
        cond4 = df['nationality'] == 'Argentina'
        df_argentina = df[cond4] 
        df_argentina.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_argentina.head(10)
    elif x == 'France':
        global cond5 
        global df_france
        cond5 = df['nationality'] == 'France'
        df_france = df[cond5]
        df_france.sort_values(by = 'overall', ascending = False, inplace = True)
        return df_france.head(10)
```

```python
#Top 5 nations that have players represent in Fifa21 using bar chart
df_top5_nation = df[['nationality']].value_counts().to_frame()
df_top5_nation.columns = ['count']
df_top5_nation.head(5).plot(kind = 'bar', figsize = (10, 6), color = 'r')
plt.title('Top 5 nations that the players represent in Fifa21')
plt.xlabel('Nations')
plt.ylabel('Number of players')
plt.show()
```

```python
#Age distribution of the players in Fifa21 using histogram
count, bin_edges = np.histogram(df['age'], 10)
df['age'].plot(kind = 'hist', figsize = (10, 6), color = 'r', xticks = bin_edges, bins = 10)
plt.title('Age distribution of the players in Fifa21')
plt.xlabel('Age range')
plt.ylabel('Number of players')
plt.show()
```

```python
#Top 5 team that have most players represent in Fifa21 using pie chart
df_team = df['team'].value_counts().to_frame()
df_team.sort_values(by = 'team', ascending = False, inplace = True)
df_team.head(5).plot(kind = 'pie', 
                     figsize = (10, 6),
                     autopct = '%1.1f%%',
                     startangle = 90,
                     shadow = True,
                     labels = None,
                     pctdistance = 1.12,
                     explode = [0.0, 0.2, 0.1, 0.1, 0.1],
                     subplots = True)
plt.title('Top 5 team that have most players represent in Fifa21')
plt.axis('equal')
plt.legend(labels = df_team.head(5).index, loc = 'upper left')
plt.show()
```

```python
#Correlation between age, overall and potential
sns.heatmap(df[['age', 'overall', 'potential']].corr(), annot = True)
plt.title('Correlation between age, overall and potential')
plt.show()
```

```python
#Compare overall with potential using bubble plot
df_compare = df.loc[:, ['age', 'overall', 'potential']]
df_overall = df['overall']/df['overall'].max()
df_potential = df['potential']/df['potential'].max()
ax0 = df_compare.plot(kind = 'scatter',
                      figsize = (30, 10),
                      x = 'age',
                      y = 'overall',
                      s = df_overall*2000+10,
                      alpha = 0.5,
                      color = 'r')
ax1 = df_compare.plot(kind = 'scatter',
                      x = 'age',
                      y = 'potential', 
                      s = df_potential*2000+10,
                      alpha = 0.35,
                      color = 'blue',
                      ax = ax0)
plt.title('Compare overall with potential')
plt.xlabel('Age')
plt.ylabel('Rating')
plt.show()
```

```python
#Compare average potential and average overall using scatter plot
df_compare = df.loc[:, ['age', 'overall', 'potential']].groupby('age').mean()
df_compare.plot(kind = 'scatter', x = 'potential', y = 'overall', figsize = (8, 6), color = 'r')
plt.title('Compare average potential and average overall')
plt.xlabel('Average Potential')
plt.ylabel('Average Overall')
plt.show()
```

```python
#Compare average potential and average overall using regression plot
df_compare = df.loc[:, ['age', 'overall', 'potential']].groupby('age').mean()
sns.regplot(x = 'potential', y = 'overall', data = df_compare, color = 'r')
plt.title('Compare average potential and average overall')
plt.xlabel('Average Potential')
plt.ylabel('Average Overall')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[['age']], df[['overall']], test_size = 0.3, random_state = 0)
lr = LinearRegression()
lr.fit(x_train, y_train)
Yhat = lr.predict(x_test)
Yhat
```

```python
ax0 = sns.distplot(df[['overall']], hist = False, label = 'Actual Values', color = 'r')
sns.distplot(Yhat, hist = False, label = 'Fitted Values', color = 'b', ax = ax0)
```

```python
x_train2, x_test2, y_train2, y_test2 = train_test_split(df[['age']], df[['potential']], test_size = 0.3, random_state = 0)
lr2 = LinearRegression()
lr2.fit(x_train2, y_train2)
Yhat2 = lr2.predict(x_test2)
```
