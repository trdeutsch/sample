```python
import pandas as pd
import requests
import numpy as np
from bs4 import BeautifulSoup
```

```python
link = " https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(link).text
```

```python
soup = BeautifulSoup(html_data,"html.parser")
```

```python
tesla_revenue = pd.DataFrame(columns=['Date','Revenue'])
for text in soup.find_all(text = 'Tesla Quarterly Revenue'):
    table = text.find_parent('table')
for row in table.find('tbody').find_all('tr'):
    col = row.find_all('td')
    date = col[0].text
    revenue = col[1].text
    tesla_revenue = tesla_revenue.append({'Date':date,'Revenue':revenue},ignore_index=True)
tesla_revenue
```

```python
tesla_revenue.replace('',np.nan,inplace = True)
tesla_revenue.tail()
```

```python
tesla_revenue.dropna(subset = ['Revenue'], axis = 0, inplace = True)
tesla_revenue.tail()
```

```python
tesla_revenue.reset_index(inplace = True)
tesla_revenue.tail()
```
