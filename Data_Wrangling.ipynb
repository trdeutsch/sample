```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
df = pd.read_csv(filename, header = None)
df.head(5)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns = headers
df.head()
```

```python
df.replace('?', np.nan, inplace = True)
df.head()
```

```python
missing_value = df.isnull()
missing_value.head()
#"True" means the value is a missing value while "False" means the value is not a missing value.
```

```python
avg_nor_loss = df["normalized-losses"].astype(float).mean(axis = 0)
avg_bore = df['bore'].astype(float).mean(axis = 0)
avg_stroke= df['stroke'].astype(float).mean(axis = 0)     
avg_peak_rpm = df['peak-rpm'].astype(float).mean(axis = 0)    
#Tính mean của các cột                
```

```python
df['normalized-losses'].replace(np.nan, avg_nor_loss, inplace = True)
df['bore'].replace(np.nan, avg_bore, inplace = True)
df['stroke'].replace(np.nan, avg_stroke, inplace = True)
df['peak-rpm'].replace(np.nan, avg_peak_rpm, inplace = True)
df.head()
#Thay những giá trị NaN tại các cột bằng giá trị mean vừa tình được
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace = True)
#Thay giá trị NaN trong cột bằng giá trị 'four' do giá trị này xuất hiện nhiều nhát 
```

```python
df.dropna(subset = ['price'], axis = 0, inplace = True)
#Xóa dòng có giá trị NaN tại cột 'price'
```

```python
df.dtypes
```

```python
df['normalized-losses'] = df['normalized-losses'].astype(float)
df['bore'] = df['bore'].astype(float)
df['stroke'] = df['stroke'].astype(float)
df['peak-rpm'] = df['peak-rpm'].astype(float)
df['price'] = df['price'].astype(float)
df['horsepower'] = df['horsepower'].astype(float)
#Format lại định dạng của các cột
```

```python
df['city-L/100km'] = 235/df['city-mpg']
df['highway-L/100km'] = 235/df['highway-mpg']
df.drop('city-mpg', axis = 1, inplace = True)
df.drop('highway-mpg', axis = 1, inplace = True)
```

```python
df['height'] = df['height']/df['height'].max()
df['width'] = df['width']/df['width'].max()
df['length'] = df['length']/df['length'].max()
```

```python
df1 = pd.get_dummies(df['fuel-type'])
df1.rename(columns = {'diesel':'fuel-diesel', 'gas':'fuel-gas'}, inplace = True)
df.drop('fuel-type', axis = 1, inplace = True)
df = pd.concat([df, df1], axis = 1)
df
```

```python
df2  = pd.get_dummies(df['aspiration'])
df2.rename(columns = {'std':'aspiration - std', 'turbo' : 'asipartion - turbo'}, inplace = True)
df.drop('aspiration', axis = 1, inplace = True)
df = pd.concat([df, df2], axis = 1)
df.head()
```

```python
df.dtypes
```

```python
bin = np.linspace(min(df['price']), max(df['price']), 4)
names = ['Low', 'Medium', 'High']
df['price-bin'] = pd.cut(df['price'], bin, labels = names, include_lowest = True)
```

```python
df['price-bin'].value_counts()
```

```python
%matplotlib inline
x = names
y = df['price-bin'].value_counts()
plt.xlabel('Price')
plt.ylabel('Count')
plt.bar(x, y)
plt.show()
```
